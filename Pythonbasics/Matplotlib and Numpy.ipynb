```python
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
```

```python
# internally displayed in jupyter notebook, no need of calling plt.show again to display
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Graph in 2D")
plt.savefig('Test.png')
```

```python
y = x*x
y
```

```python
plt.plot(x,y,'r+',linestyle='dashed',linewidth=2,markersize=10)
```

```python
#creating subplot

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(y,x,'g--')
plt.subplot(2,2,3)
plt.plot(y,x,'b--')
plt.subplot(2,2,4)
plt.plot(y,x,'g-')
```

```python
#barplot

x1=[2,3,4,5]
y1=[11,32,54,63]

x2=[3,5,7,9]
y2=[12,23,34,45]
plt.bar(x1,y1,color='g')
plt.bar(x2,y2)
```

```python
#histogram

a = np.array([12,34,21,43,23,65,78])
plt.hist(a)
```

```python
# boxplot 

data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data)
```

```python
# Piechart
# Data to plot

labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, .2, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
