# QQ plot

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
```

```python
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample,kde=True)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(log_normal_data)
```

```python
plot_data(np.log(log_normal_data))
```

```python
p=0.3
sample_size=1000

bernouli_data=np.random.choice([0,1],size=sample_size,p=[1-p,p])

sns.histplot(bernouli_data,bins=2,kde=False,color="blue")
```

```python
no_of_trials=10
prob=0.6

binomial_data=np.random.binomial(no_of_trials,prob,1000)
```

```python
sns.distplot(binomial_data)
```

```python
poisson_data=np.random.poisson(lam=5, size=1000)

ax = sns.distplot(poisson_data,kde=False,color='blue')
ax.set(xlabel='Poisson distribution', ylabel='Frequency')

plt.show()
```
